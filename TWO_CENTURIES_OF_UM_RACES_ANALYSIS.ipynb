```python
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.contains('USA', na=False))  ]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
#remove USA from event name
```

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
# clean up athlete age
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#remove h from athlete performance 
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

```python
# drop colums : athlete club, athlete country , athlete year of birth , athlete age category 
```

```python
df2 = df2.drop(['Athlete club' ,'Athlete country' , 'Athlete year of birth' , 'Athlete age category'], axis =1)
```

```python
df2.head()
```

```python
# clean up null values 
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#check for dupes 
```

```python
df2[df2.duplicated() == True]
```

```python
# reset index
```

```python
df2.reset_index(drop = True)
```

```python
#fix types 
```

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
#rename columns
```

```python
#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#athlete_age                    int32
#dtype: object
```

```python
df2 = df2.rename(columns = {
    'Year of event' : 'year',
    'Event dates' : 'race_day' ,
    'Event name' : 'race_name',
    'Event distance/length' : 'race_lenght' ,
    'Event number of finishers' : 'race_number_of_finisher' ,
    'Athlete performance' : 'athlete_performance',
    'Athlete gender' : 'athlete_gender',
    'Athlete average speed' : 'athlete_average_speed',
    'Athlete ID' : 'athlete_id' 
})
```

```python
#charts and graphs
```

```python
sns.histplot(df2['race_lenght'])
```

```python
sns.histplot(df2, x= 'race_lenght', hue = 'athlete_gender')
```

```python
sns.displot(df2[df2['race_lenght'] == '50mi'] ['athlete_average_speed'])
```

```python
sns.violinplot(data = df2, x='race_lenght', y='athlete_average_speed', hue='athlete_gender', split= True, inner= 'quart' )
```

```python
sns.lmplot(data=df2, x= 'athlete_age', y='athlete_average_speed', hue= 'athlete_gender')
```

```python
#Difference in speed for the 50k, 50 mi male to female
```

```python
df2.groupby(['race_lenght', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
#what age groups are the best in the 50m race (20 + races min)
```

```python
df2.query('race_lenght == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False).query('count>19')
```
